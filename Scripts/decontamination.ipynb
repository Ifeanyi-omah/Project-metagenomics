```python
from ete3 import NCBITaxa
import pandas as pd

# Load the data
lca = pd.read_csv('/Volumes/aine_store/SENZOR_project/SPlited_SENSOR_porject/IDseq/Combined_SENZOR_results/contig_stats_lca.tsv', sep='\t')

# Initialize NCBI Taxa instance
ncbi = NCBITaxa()

# Function to retrieve taxonomic ranks
def get_taxonomy_ranks(tax_id, ranks=('phylum', 'family')):
    try:
        lineage = ncbi.get_lineage(tax_id)
        names = ncbi.get_taxid_translator(lineage)
        ranks_info = ncbi.get_rank(lineage)
        result = {rank: names[tax] for tax, rank in ranks_info.items() if rank in ranks}
        return [result.get(rank, 'NA') for rank in ranks]
    except Exception as e:
        return ['NA'] * len(ranks)

# Add phylum and family columns
lca[['phylum', 'family']] = lca['tax_id'].apply(lambda x: pd.Series(get_taxonomy_ranks(int(x))))

# Save the updated file
output_path = '/Volumes/aine_store/SENZOR_project/SPlited_SENSOR_porject/IDseq/Combined_SENZOR_results/contig_stats_lca_updated.tsv'
lca.to_csv(output_path, sep='\t', index=False)
```

```python
sample_table = pd.read_csv('/Volumes/aine_store/SENZOR_project/SPlited_SENSOR_porject/IDseq/Combined_SENZOR_results/correct_Idseq_metatda.csv')

sample_table['water'] = sample_table['water'].fillna(False).astype(bool)
print(sample_table['water'].dtype)  # should now be bool
```

```python
# Mark as water if sample starts with one of these prefixes
sample_table['water'] = sample_table['sample'].str.startswith(('SSP_', 'NTC_', 'NE_'))

# Keep only relevant columns
sample_table = sample_table[['sample',
                             'total_reads',
                             'nonhost_reads',
                             'total_ercc_reads',
                             'compression_ratio',
                             'water']]

# Compute extra columns
sample_table['input_conc'] = sample_table.eval('(total_reads - total_ercc_reads) / total_ercc_reads')
sample_table['nonhost_frac'] = sample_table.eval('nonhost_reads * compression_ratio / total_reads')

# Count how many are marked as water
n_water = sample_table['water'].sum()
n_water
```

```python
def sum_over(df, category):
    return df[[category, 'sample', 'read_count']].groupby([category, 'sample']).sum().reset_index()
```

```python
read_counts = sum_over(contig_calls, 'poly_group')
read_counts
```

```python
def decontam(read_counts, category, fdr):
    background_levels = (read_counts.merge(sample_table[sample_table['water']],
                                          on='sample', validate='many_to_one')
                         .eval('r_per_ercc = read_count*compression_ratio/total_ercc_reads')
                         [['r_per_ercc', category]]
                         .groupby(category).sum()/n_water)
    
    df = read_counts.merge(sample_table, on = 'sample', how = 'left')
    df = df[~df['water']]
    df = df.merge(background_levels, on=category, how = 'left').fillna({'r_per_ercc': 0}).eval(
    'background_reads = total_ercc_reads*r_per_ercc/compression_ratio')
    df['is_contam'] = df['read_count'] < df['background_reads']/fdr
    
    return df[['sample', category, 'is_contam', 'read_count']]
```

```python
from ete3 import NCBITaxa
ncbi = NCBITaxa()
def taxid2name(taxid):
    return ncbi.get_taxid_translator([taxid])[taxid]
```

```python
# Updated partition list to include mammals
partition = "Apicomplexa Crithidia Pecora Carnivora Homininae Rodentia Leporidae Aves Platyhelminthes Nematoda Fungi Mammalia".split()

# Get name translator for the updated partition
partition = ncbi.get_name_translator(partition)

# Reverse the dictionary for quick lookup
partition = {v[0]: k for k, v in partition.items()}

def get_category(taxid):
    if not taxid:
        return None
    try:
        lineage = ncbi.get_lineage(taxid)
    except:
        return 'NA'
    for k in partition:
        if k in lineage:
            return partition[k]
    else:
        return 'NA'
```
