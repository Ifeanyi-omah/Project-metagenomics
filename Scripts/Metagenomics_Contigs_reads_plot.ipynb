```python
import os
import pandas as pd

# Corrected directory path
directory_path = "/Volumes/aine_store/SENZOR_project/SPlited_SENSOR_porject/IDseq/chicken_16386_Contigs_summary/"

# Define the columns to retain
columns_to_retain = ["contig_name", "read_count", "contig_length", "contig_coverage"]

# List to store processed dataframes
processed_dfs = []

# Process each file in the directory
for file_name in os.listdir(directory_path):
    if file_name.endswith("_contig_summary_report.csv"):
        file_path = os.path.join(directory_path, file_name)
        # Extract sample name based on the format you specified
        sample_name = "_".join(file_name.split("_")[:3])
        
        # Load the CSV file
        try:
            df = pd.read_csv(file_path)
            
            # Filter the columns
            filtered_df = df[columns_to_retain]
            
            # Add the sample name column
            filtered_df["sample_name"] = sample_name
            
            # Save the processed dataframe
            processed_dfs.append(filtered_df)
        except Exception as e:
            print(f"Error processing {file_name}: {e}")

# Combine all processed dataframes into one
if processed_dfs:
    combined_df = pd.concat(processed_dfs, ignore_index=True)

    # Save the combined dataframe to a new JSON file
    output_json_path = os.path.join(directory_path, "filtered_combined_contig_summary.json")
    combined_df.to_json(output_json_path, orient="records", lines=True)
    print(f"Filtered data saved to {output_json_path}")
else:
    print("No files were processed.")
```

```python
import matplotlib.pyplot as plt

# Ensure the directory path is correct or update as needed
directory_path = "/Volumes/aine_store/SENZOR_project/SPlited_SENSOR_porject/IDseq/chicken_16386_Contigs_summary/"
output_file = os.path.join(directory_path, "filtered_combined_contig_summary.csv")

# Check if the filtered file exists
if os.path.exists(output_file):
    # Load the filtered combined file
    filtered_df = pd.read_csv(output_file)
    
    # Calculate total reads
    total_reads = filtered_df["read_count"].sum()
    print(f'total reads:{total_reads}')
    
    # Calculate total contigs
    total_contigs = filtered_df["contig_name"].nunique()
    print(f'total config:{total_contigs}')
    
    # Calculate total contigs per sample
    contigs_per_sample = filtered_df.groupby("sample_name")["contig_name"].count()
    
    # Plot total contigs per sample
    plt.figure(figsize=(20, 8))
    contigs_per_sample.sort_values().plot(kind="bar", title="Non-host Contigs Per Sample")
    plt.xlabel("Sample Name")
    plt.ylabel("Number of Contigs")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
    
    # Output results
    {
        "Total Reads": total_reads,
        "Total Contigs": total_contigs
    }
else:
    "Filtered combined file not found. Please ensure the processing step completed successfully."
```
